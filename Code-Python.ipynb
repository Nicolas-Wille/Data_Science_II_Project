Create a function that: 
1) Takes as input (i) the initial EDX dataset, and (ii) the path of the csv dataset containing the raster informations. 
2) First transforms the csv file by keeping only the id and the value (here the mean)
3) Then, renames the mean with the name of the csv
4) Merge it with our initial dataset df using the index "Field1"

```python
def path_leaf(path):
    head, tail = ntpath.split(path)
    return tail or ntpath.basename(head)

def MyFunction(path, df):
    df_Raster = pd.read_csv(path, sep=";")
    name_path = path_leaf(path).split('.csv')[0]
    df_Mean = df_Raster[["Field1","MEAN"]]
    df_Mean_Renamed = df_Mean.rename({"MEAN":name_path}, axis = 1)
    df = pd.merge(df, df_Mean_Renamed, on='Field1', how='outer')
    return df
```
